```python
df.tpep_pickup_datetime=pd.to_datetime(df.tpep_pickup_datetime)
df.tpep_dropoff_datetime=pd.to_datetime(df.tpep_dropoff_datetime)

```

```python
df=df.drop_duplicates().reset_index(drop=True)
df['trip_id']=df.index
```

```python
df.head()
```

```python
datetime_dt=df[['tpep_pickup_datetime','tpep_dropoff_datetime']].reset_index(drop=True)
datetime_dt['pick_hour']=datetime_dt['tpep_pickup_datetime'].dt.hour
datetime_dt['pick_day']=datetime_dt['tpep_pickup_datetime'].dt.day
datetime_dt['pick_month']=datetime_dt['tpep_pickup_datetime'].dt.month
datetime_dt['pick_year']=datetime_dt['tpep_pickup_datetime'].dt.year

datetime_dt['drop_hour']=datetime_dt['tpep_dropoff_datetime'].dt.hour
datetime_dt['drop_day']=datetime_dt['tpep_dropoff_datetime'].dt.day
datetime_dt['drop_month']=datetime_dt['tpep_dropoff_datetime'].dt.month
datetime_dt['drop_year']=datetime_dt['tpep_dropoff_datetime'].dt.year
```

```python
datetime_dt['datetime_id']=datetime_dt.index
```

```python
datetime_dt=datetime_dt[['datetime_id','tpep_pickup_datetime','pick_hour','pick_day','pick_month','pick_year','tpep_dropoff_datetime','drop_hour','drop_day','drop_month','drop_year']]
```

```python
datetime_dt
```

```python
passenger_cnt_dt=df[['passenger_count']].reset_index(drop=True)
passenger_cnt_dt['passenger_count_id']=passenger_cnt_dt.index
passenger_cnt_dt=passenger_cnt_dt[['passenger_count_id','passenger_count']]

trip_distance_dt=df[['trip_distance']].reset_index(drop=True)
trip_distance_dt['trip_distance_id']=trip_distance_dt.index
trip_distance_dt=trip_distance_dt[['trip_distance_id','trip_distance']]
```

```python
rate_code_type={
    1:'Standard Rate',
    2:'JFK',
    3:'Newark',
    4:'Nassau or Westchester',
    5:'Negotiated Fare',
    6:'Group Ride'
}

rate_code_dt=df[['RatecodeID']].reset_index(drop=True)
rate_code_dt['rate_code_id']=rate_code_dt.index
rate_code_dt['rate_code_name']=rate_code_dt['RatecodeID'].map(rate_code_type)
rate_code_dt=rate_code_dt[['rate_code_id','RatecodeID','rate_code_name']]
```

```python
rate_code_dt.head()
```

```python
pickup_loc_dt = df[['pickup_longitude', 'pickup_latitude']].reset_index(drop=True)
pickup_loc_dt['pickup_loc_id'] = pickup_loc_dt.index
pickup_loc_dt = pickup_loc_dt[['pickup_loc_id','pickup_latitude','pickup_longitude']] 


dropoff_loc_dt = df[['dropoff_longitude', 'dropoff_latitude']].reset_index(drop=True)
dropoff_loc_dt['dropoff_loc_id'] = dropoff_loc_dt.index
dropoff_loc_dt = dropoff_loc_dt[['dropoff_loc_id','dropoff_latitude','dropoff_longitude']]
```

```python
payment_type_name = {
    1:"Credit card",
    2:"Cash",
    3:"No charge",
    4:"Dispute",
    5:"Unknown",
    6:"Voided trip"
}
payment_type_dt = df[['payment_type']].reset_index(drop=True)
payment_type_dt['payment_type_id'] = payment_type_dt.index
payment_type_dt['payment_type_name'] = payment_type_dt['payment_type'].map(payment_type_name)
payment_type_dt = payment_type_dt[['payment_type_id','payment_type','payment_type_name']]
```

```python
fact_table = df.merge(passenger_cnt_dt, left_on='trip_id', right_on='passenger_count_id') \
             .merge(trip_distance_dt, left_on='trip_id', right_on='trip_distance_id') \
             .merge(rate_code_dt, left_on='trip_id', right_on='rate_code_id') \
             .merge(pickup_loc_dt, left_on='trip_id', right_on='pickup_loc_id') \
             .merge(dropoff_loc_dt, left_on='trip_id', right_on='dropoff_loc_id')\
             .merge(datetime_dt, left_on='trip_id', right_on='datetime_id') \
             .merge(payment_type_dt, left_on='trip_id', right_on='payment_type_id') \
             [['trip_id','VendorID', 'datetime_id', 'passenger_count_id',
               'trip_distance_id', 'rate_code_id', 'store_and_fwd_flag', 'pickup_loc_id', 'dropoff_loc_id',
               'payment_type_id', 'fare_amount', 'extra', 'mta_tax', 'tip_amount', 'tolls_amount',
               'improvement_surcharge', 'total_amount']]
```
